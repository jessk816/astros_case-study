Breakout hitter Lou Rice is off to a tremendous start to the season, with hits in 43 of his first 100 at-bats. Going into the season our belief about Lou’s true-talent batting average, i.e., what Lou’s hits per at-bat rate would be in the limit, could be described as a Beta distribution parameterized with α = 58 and β = 142. Given our beliefs at the outset of the season and Lou’s performance through his first 100 at-bats, what should we believe Lou’s chance is of getting at least 200 hits in a total of 500 at-bats? Describe your reasoning.

## Imports

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import beta
```

## Beta Distribution Plot

```python
# set shape paremeters
a, b = 58, 142

# generate value
x = np.linspace(beta.ppf(0.01, a, b),
                beta.ppf(0.99, a, b), 100)

# plot beta distribution
plt.figure(figsize=(7,7))
plt.plot(x, beta.pdf(x, a, b), 'b-')
plt.title('Beta Distribution', fontsize='15')
plt.xlabel('Values of Random Variable X (0, 1)', fontsize='15')
plt.ylabel('Probability', fontsize='15')

plt.show()
```

```python
# set shape paremeters
a, b = 43, 100              # actual
a1, b1 = 58, 142            # predicted 
a2, b2 = 58+200, 142+300    # test

# generate value
x = np.linspace(beta.ppf(0.01, a, b),
                beta.ppf(0.99, a, b), 100)
x1 = np.linspace(beta.ppf(0.01, a1, b1),
                beta.ppf(0.99, a1, b1), 100)
x2 = np.linspace(beta.ppf(0.01, a2, b2),
                beta.ppf(0.99, a2, b2), 100)

# plot beta distribution
plt.figure(figsize=(7,7))
plt.plot(x, beta.pdf(x, a, b), 'g-', label='Actual')
plt.plot(x1, beta.pdf(x1, a1, b1), 'b-', label='Predicted')
plt.plot(x2, beta.pdf(x2, a2, b2), 'r-', label='Test')

plt.legend()
plt.title('Beta Distribution', fontsize='15')
plt.xlabel('Values of Random Variable X (0, 1)', fontsize='12')
plt.ylabel('Probability', fontsize='12')

plt.show()
```
